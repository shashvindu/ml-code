```python
import pandas as pd
```

```python
df=pd.read_csv("titanic_train.csv")
df.head()
```

```python
target=df.Survived
target
```

```python
newdf=df.drop(['PassengerId','Survived','Parch','Name','SibSp','Ticket','Cabin','Embarked'],axis='columns')
newdf.head()
```

```python
dumm=pd.get_dummies(newdf.Sex)
dumm.head()
```

```python
 df_fin=pd.concat([newdf,dumm],axis='columns')
df_fin.head()
```

```python
df_n=df_fin.drop(['Sex'],axis='columns')
df_n.head()
```

```python
df_n.info()
```

```python
df_n.head(20)
```

```python
import numpy as np
df_n['Age']=np.floor(df_n.Age.fillna(df_n.Age.mean()))
df_n.head(10)
```

```python
df_n.describe()
```

```python
df_n.groupby('male').mean()
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(df_n,target,test_size=0.2)
```

```python
xtrain.head(10)
```

```python
ytrain.head(10)
```

```python
from sklearn.naive_bayes import GaussianNB
nb=GaussianNB()
nb
```

```python
nb.fit(xtrain,ytrain)
```

```python
nb.predict(xtest)[:10]
```

```python
ytest.head(10)
```

```python
nb.score(xtest,ytest)
```

```python
nb.predict_proba(xtest.head(10))
```
