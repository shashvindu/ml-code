```python
import numpy as np
```

```python
def gradient_descent(x,y):
    m_curr=b_curr=0
    itr=1000
    n=len(x)
    lr=0.08
    plt.scatter(x,y,color='red',marker='+',linewidth='5')
    for i in range(itr):
        y_pred=m_curr*x+b_curr
        c=(1/n)*sum([val**2 for val in(y-y_pred)])
        md=-(2/n)*sum(x*(y-y_pred))
        bd=-(2/n)*sum(y-y_pred)
        m_curr=m_curr-lr*md
        b_curr=b_curr-lr*bd
        plt.plot(x,y_pred,color='green')
        print("m: {}, b: {}, cost: {}, iteration: {}".format(m_curr,b_curr,c,i))

x=np.array([1,2,3,4,5])
y=np.array([5,7,9,11,13])

gradient_descent(x,y)
```
